# On the definition of u

A constraint is placed on the value of $u$ just below Eq. (4.7) in Lebedev. The constraint is based on the phase advances being positive and proportional to $(1 - u)$ if there is no longitudinal field, therefore $u$ must be less than or equal to 1. There is no constraint on the sign of u. The velocity vectors along the beam envelope in real space are shown in the figure below as $u$ is changed.

```python
Image(filename='negative_u.png') 
```

Basically the $\langle{xy'}\rangle$ and $\langle{yx'}\rangle$ moments now have the same sign, so the beam is kind of "skewing" instead of rotating. A similar thing happens in the opposite direction when $u > 1$. Interestingly, althought the intrinsic emittance is still the same, the condition $\varepsilon_1 = \varepsilon_x + \varepsilon_y$ is no longer true, but instead $\varepsilon_1 = \varepsilon_x - \varepsilon_y$. It kind of reminded me of the fact that a 2D disc can have infinite area but still have zero 3D volume. It is a bit weird that the intrinsic emittance does not change at all no matter how large the apparent emittances get.

Looking at the matched beam covariance matrix in the appendix of Lebedev and setting $\varepsilon_2 = 0$ in Eqs. (A.1.1 - A.1.3) gives

\begin{align}
    \left(\varepsilon_x / \varepsilon_1\right)^2 &= (1 - u)^2, \\
    \varepsilon_x / \varepsilon_1 &= \pm \sqrt{(1 - u)^2}, \\
    \varepsilon_x / \varepsilon_1 &= \sqrt{(1 - u)^2}, \\
    \varepsilon_x / \varepsilon_1 &= 1 - u.
\end{align}

The $\pm$ sign is eliminated because the emittance must be greater than or equal to zero. Similarly for the $y$ direction:

\begin{align}
    \left(\varepsilon_y / \varepsilon_1\right)^2 &= u^2, \\
    \varepsilon_y / \varepsilon_1 &= \pm \sqrt{u^2}, \\
    \varepsilon_y / \varepsilon_1 &= \sqrt{u^2}, \\
    \varepsilon_y / \varepsilon_1 &= u.
\end{align}

Finally,

$$\frac{\left(\varepsilon_x + \varepsilon_y\right)}{\varepsilon_1} = \left(1 - u\right) + u = 1,$$ 

but, from the figure above, this is not true when $u$ is not in the range [0, 1]. So it seems we can rule out any solutions for which $u$ is not between zero and one on the grounds that they cannot be matched to a linear/stable lattice. I tested this out by taking an uncoupled FODO lattice, without space charge, and creating a matched beam with u=0.5 ($\varepsilon_x = \varepsilon_y$). Then I changed the sign of $u$ while keeping everything else the same. The resulting beam is no longer matched.
